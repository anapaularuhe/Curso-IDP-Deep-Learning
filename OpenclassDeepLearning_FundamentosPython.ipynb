### Passo 1 - Obter os dados fonte e armazenar localmente

```python
## Conectando ao repositório de dados em https://www.kaggle.com/c/dogs-vs-cats/data
## e baixando as imagens para uma pasta local

!wget --no-check-certificate \
    https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip \
    -O /tmp/cats_and_dogs_filtered.zip
```

```python
## Extraindo os arquivos baixados

import os
import zipfile

local_zip = '/tmp/cats_and_dogs_filtered.zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall('/tmp')
zip_ref.close()
```

```python
## O resultado é uma subpasta com dois grupos de imagens, respectivamente nos 
## diretórios 'train' e 'validation', os quais, por sua vez, contém os subdiretórios
## 'dogs' e 'cats'.

## Para trabalhar com esses dados, é necessário armazenar os correspondentes
## caminhos em variáveis:

base_dir = '/tmp/cats_and_dogs_filtered'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')

# Imagens de gatos para treinar a RN
train_cats_dir = os.path.join(train_dir, 'cats')

# Imagens de cães para treinar a RN
train_dogs_dir = os.path.join(train_dir, 'dogs')

# Imagens de gatos para validar a RN
validation_cats_dir = os.path.join(validation_dir, 'cats')

# Imagens de cães para validar a RN
validation_dogs_dir = os.path.join(validation_dir, 'dogs')
```

```python
## Verificando o conteúdo baixado:

train_cat_fnames = os.listdir(train_cats_dir)
print(train_cat_fnames[:10])

train_dog_fnames = os.listdir(train_dogs_dir)
train_dog_fnames.sort()
print(train_dog_fnames[:10])
```

```python
## Verificando o total de imagens disponíveis:

print('Número de imagens de gatos para treinamento:', len(os.listdir(train_cats_dir)))
print('Número de imagens de cães para treinamento:', len(os.listdir(train_dogs_dir)))
print('Número de imagens de gatos para validação:', len(os.listdir(validation_cats_dir)))
print('Número de imagens de cães para validação:', len(os.listdir(validation_dogs_dir)))
```

```python
## Será que eu posso visualizar as imagens diretamente no Python?

%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Grade de 4 linhas x 4 colunas
nrows = 4
ncols = 4

# Índice para iterar ao longo dos dados
pic_index = 0

fig = plt.gcf()
fig.set_size_inches(ncols * 4, nrows * 4)

# Exibindo as imagens em grupos de 8
n = 2
pic_index += 8 * n
next_cat_pix = [os.path.join(train_cats_dir, fname) 
                for fname in train_cat_fnames[pic_index-8:pic_index]]
next_dog_pix = [os.path.join(train_dogs_dir, fname) 
                for fname in train_dog_fnames[pic_index-8:pic_index]]

for i, img_path in enumerate(next_cat_pix+next_dog_pix):
  sp = plt.subplot(nrows, ncols, i + 1) # subplot iniciando em 1
  sp.axis('Off') # retirar eixos e grades

  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show()
```

### Passo 2 - Construir o modelo da rede neural

```python
## Carregando o Keras
from tensorflow.keras import layers
from tensorflow.keras import Model

## Parâmetros de codificação das imagens
## Serão mapeadas para 150x150 pixels e três dimensões de cores (R,G,B)
img_input = layers.Input(shape=(150, 150, 3))

## Primeira convolução extrai 16 mapas de 3x3, com max-pooling de 2x2
x = layers.Conv2D(16, 3, activation='relu')(img_input)
x = layers.MaxPooling2D(2)(x)

## Segunda convolução extrai 32 mapas de 3x3, com max-pooling de 2x2
x = layers.Conv2D(32, 3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)

## Terceira convolução extrai 64 mapas de 3x3, com max-pooling de 2x2
x = layers.Conv2D(64, 3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)
```

<img src="https://machinelearningmastery.com/wp-content/uploads/2019/01/Example-of-a-Filter-Applied-to-a-Two-Dimensional-input-to-create-a-Feature-Map.png" width="400"><br>
<img src="https://lh5.googleusercontent.com/jOohqE8T9IYqkir1OzJN2vRdeykH1yGnwEy9oWK85EWf2aP-jg10YZt_3yaVDeCZVS9713QDACk0l1iYX5e5EfE2CUqbSLVaq5UBeGL91a92v3ZMRyb4FLobxjN5pDh4RBc1pT-8" width="400">

```python
## Conectando as camadas da RN
x = layers.Flatten()(x)

## Definindo os parâmetros de ativação da primeira camada (ReLU = rectified linear activation 
## function) com 512 unidades
x = layers.Dense(512, activation='relu')(x)

## Definindo a segunda camada com ativação sigmóide
## O resultado é a probabilidade de a imagem pertencer a uma classe (0) ou outra (1)
output = layers.Dense(1, activation='sigmoid')(x)

## Criando o modelo:
## entrada = mapa de características das imagens
## saída =apa de características das imagens + camadas de convolução/maxpooling + primeira  
## camada de ativação + camada de saída sigmóide
model = Model(img_input, output)
```

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/88/Logistic-curve.svg/1280px-Logistic-curve.svg.png" width="400">

```python
## Observando os parâmetros do modelo

model.summary()
```

```python
## Configurando os parâmetros de treinamento do modelo

from tensorflow.keras.optimizers import RMSprop

## binary_crossentropy é escolhido por ser um problema de classificação binária (cães ou gatos)
model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(lr=0.001),
              metrics=['acc'])
```

### Passo 3 - Pré-processar das imagens

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Aplicando fatores de escala 1./255
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

# Fornecendo imagens de treinamento em pacotes (batch_size) de 20
train_generator = train_datagen.flow_from_directory(
        train_dir,  # diretório com as imagens para treinamento
        target_size=(150, 150),  # Todas as imagens serão reduzidas para 150 x 150 px
        batch_size=20,
        # para binary_crossentropy, necessitamos rótulos binários
        class_mode='binary')

# Fornecendo imagens de validação em pacotes (batch_size) de 20
validation_generator = val_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='binary')
```

### Passo 4 - Treinar a rede neural

```python
# Treinando o modelo por 15 épocas

history = model.fit_generator(
      train_generator,
      steps_per_epoch=100,  # 2000 imagens = batch_size=20 * steps
      epochs=15,
      validation_data=validation_generator,
      validation_steps=50,  # 1000 imagens = batch_size=20 * steps
      verbose=2)
```

### Passo 5 - Visualizar representações intermediárias

```python
import numpy as np
import random
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.applications.resnet50 import ResNet50,preprocess_input, decode_predictions

# Definindo um modelo de visualização
successive_outputs = [layer.output for layer in model.layers[1:]]
visualization_model = Model(img_input, successive_outputs)

# Selecionando uma imagem aleatória
cat_img_files = [os.path.join(train_cats_dir, f) for f in train_cat_fnames]
dog_img_files = [os.path.join(train_dogs_dir, f) for f in train_dog_fnames]
img_path = random.choice(cat_img_files + dog_img_files)

img = load_img(img_path, target_size=(150, 150))  # carregando a imagem
x = img_to_array(img)  # convertendo para array numérico (150, 150, 3)
x = x.reshape((1,) + x.shape)  # reordenando o array (1, 150, 150, 3)

# Redefinindo o fator de escala 1/255
x /= 255

# Passando a imagem pela rede neural
successive_feature_maps = visualization_model.predict(x)

# Nomes das camadas
layer_names = [layer.name for layer in model.layers[1:]]

# Convertendo para imagens novamente
for layer_name, feature_map in zip(layer_names, successive_feature_maps):
  if len(feature_map.shape) == 4:
      n_features = feature_map.shape[-1] 
      size = feature_map.shape[1]
      display_grid = np.zeros((size, size * n_features))
      for i in range(n_features):
        x = feature_map[0, :, :, i]
        x -= x.mean()
        x /= x.std()
        x *= 64
        x += 128
        x = np.clip(x, 0, 255).astype('uint8')
        display_grid[:, i * size : (i + 1) * size] = x
      scale = 20. / n_features
      plt.figure(figsize=(scale * n_features, scale))
      plt.title(layer_name)
      plt.grid(False)
      plt.imshow(display_grid, aspect='auto', cmap='viridis')
```

### Passo 6 - Testar a precisão do resultados

```python
# Obtendo histórico de resultados de treinamento e validação para cada época
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

# Total de épocas
epochs = range(len(acc))

# Plotando acurácia de treinamento e validação para cada época
plt.plot(epochs, acc)
plt.plot(epochs, val_acc)
plt.title('Acurácia de treinamento e validação')

plt.figure()

# Plotando perdas de treinamento e validação para cada época
plt.plot(epochs, loss)
plt.plot(epochs, val_loss)
plt.title('Perdas de treinamento e validação')
```
